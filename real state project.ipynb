## Scikit-learn Design

# primarily 3 types of objects

# 1.Estimators- it estimates some parameter based on data set . Ex: imputer
-->it has fit and transform method

# 2. Transformers 
-->takes input return output fit theke amra ja paise tar upor vitti kore
--> it has convenience function called fit_transform() 
-->fit and transform ekshathe kaj kore
->>it is optimize

# 3. Predictors : ex: linear regression
-->fit() and predict() two function
-->score() function ase ja predict korbe evaluate ki rokom hoise


# Feature Scaling

scaling korle data 0-1 range er moddhe ase . tasara data range onke beshi hoy dekhe graph e plot kora osubidha hoye jay.

1.Min-max scaling(Normalization)

(value-min)/(max-min)

2.standardization

(value-mean)/standardDeviation

-->sklearn provide StandardScaler for this


# CREATing a Pipeline

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

my_pipeline= Pipeline([
    
#     impute korsi and seita abar standartScale korsi ei function e
    ('imputer',SimpleImputer(strategy='median')),
    ('std_scaler',StandardScaler()),
])


```

```python
housing_num_tr=my_pipeline.fit_transform(housing)
len(housing_num_tr)
```

# Using better Evalution techinique Cross-Validation

```python
# let we have 5 data set 1 2 3 4 5
# first 2-4 data niye train korbo 1 diye test korbo
# then 1 3 4 5 diye train krobo 2 diye test korbo
# then 1 2 4 5 diye train ...... 3 diye test korbo
#........

from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)

# score negative asbe eijonno - diye + korlam

rmse_scores= np.sqrt(-scores)



```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores",scores)
    print("Mean:  ",scores.mean())
    print("Standart deviation : ",scores.std())
```

```python
print_scores(rmse_scores)
```
